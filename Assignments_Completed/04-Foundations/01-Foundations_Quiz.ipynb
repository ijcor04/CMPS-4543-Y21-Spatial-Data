## Q4. File I/O
Write Python code that saves `Hello World` to a file called `hello.txt`, then reads it back and prints the contents.

```python
with open('hello.txt', 'w') as f:
    f.write('Hello, World!')
with open('hello.txt', 'r') as g:
    content = g.read()
print(content)
```

## Q6. Plotting
Create a line plot of $y = x^2$ for $x$ values from 0 to 10.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y = x**2
plt.plot(x, y)
plt.xlabel('X-values')
plt.ylabel('Y-values (X squared)')
plt.title('Simple Line Plot')
plt.grid(True)
plt.show()
```

## Q9. Timing
Compare the runtime of squaring numbers using a Python loop vs. NumPy. Use `%timeit` for each.

```python
import numpy as np

data = np.random.rand(1000)
py_list = data.tolist()
%timeit [x**2 for x in py_list]
%timeit data**2
```
